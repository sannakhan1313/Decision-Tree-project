```python
df = load_breast_cancer()
```

```python
import pandas as pd
Data = pd.DataFrame(df['data'])
Data.columns=(['mean radius', 'mean texture', 'mean perimeter', 'mean area',
        'mean smoothness', 'mean compactness', 'mean concavity',
        'mean concave points', 'mean symmetry', 'mean fractal dimension',
        'radius error', 'texture error', 'perimeter error', 'area error',
        'smoothness error', 'compactness error', 'concavity error',
        'concave points error', 'symmetry error',
        'fractal dimension error', 'worst radius', 'worst texture',
        'worst perimeter', 'worst area', 'worst smoothness',
        'worst compactness', 'worst concavity', 'worst concave points',
        'worst symmetry', 'worst fractal dimension'])
```

```python
Data['Target']=df['target']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(25,15))
sns.heatmap(Data.corr(),annot=True)
```

```python
import numpy as np
```

```python
Y = Data.iloc[:,-1].values
X = Data.iloc[:,:-1].values
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_new = sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X_new,Y,test_size=0.3,random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion='entropy',ccp_alpha=0.0,random_state=1)
dt.fit(X_train,Y_train)
```

```python
Y_pred = dt.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
confusion_matrix(Y_test,Y_pred)
```

```python
print(classification_report(Y_test,Y_pred))
```

```python
dt.score(X_train,Y_train)
```

```python
dt.score(X_test,Y_test)
```

## Pruning

```python
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(20,15))
tree.plot_tree(dt)
```

```python
data1=dt.cost_complexity_pruning_path(X_train,Y_train)
ccp_alphas,impurities=data1.ccp_alphas,data1.impurities
```

```python
ccp_alphas
```

```python
impurities
```

```python
treee=[]
for i in ccp_alphas:
    dt = DecisionTreeClassifier(random_state=1,ccp_alpha=i)
    dt.fit(X_train,Y_train)
    treee.append(dt)
print('no of nodes in the last tree',treee[-1].tree_.node_count)
print('value of ccpalpha is',ccp_alphas[-1])
```

```python
fig,ax=plt.subplots()
accuracy=[]
acc_test=[]
for i in treee:
    accuracy.append(i.score(X_train,Y_train))
    acc_test.append(i.score(X_test,Y_test))
ax.plot(ccp_alphas[:-1],accuracy[:-1],marker='o',drawstyle='steps-post')
ax.plot(ccp_alphas[:-1],acc_test[:-1],marker='o',drawstyle='steps-post')
plt.grid()
```

```python
#final model with ccpalpha value of 0.01
dt_final = DecisionTreeClassifier(random_state=1,ccp_alpha=0.01)
dt_final.fit(X_train,Y_train)
```

```python
y_p=dt_final.predict(X_test)
```

```python
print(classification_report(Y_test,y_p))
```

```python
from sklearn import tree
tree.plot_tree(dt_final,filled=True)
```
